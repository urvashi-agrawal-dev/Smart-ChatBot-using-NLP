```python
import json

intents = {
    "intents": [
        {
            "tag": "greeting",
            "patterns": ["Hi", "Hello", "Good morning", "Hey"],
            "responses": ["Hello! How can I help you today?"]
        },
        {
            "tag": "password_reset",
            "patterns": ["I forgot my password", "reset my password", "can't login"],
            "responses": ["Click here to reset your password: [reset link]"]
        },
        {
            "tag": "goodbye",
            "patterns": ["Bye", "See you", "Goodbye"],
            "responses": ["Goodbye! Have a nice day."]
        }
    ]
}

with open("intents.json", "w") as file:
    json.dump(intents, file)
```

```python
import nltk
from nltk.stem import WordNetLemmatizer #it redices words to base form like running to run
from sklearn.feature_extraction.text import CountVectorizer #converts text into nos we needed this for ml
from sklearn.linear_model import LogisticRegression  #ml model to predict intent
import random #for picking a random response from lsit

nltk.download('punkt') #convert tokeniser for splitting text into words
nltk.download('wordnet') #used for lemmatizing

# Load data
with open("intents.json") as file:
    data = json.load(file)

lemmatizer = WordNetLemmatizer()

sentences = []
labels = []

for intent in data['intents']:
    for pattern in intent['patterns']:
        sentences.append(pattern)
        labels.append(intent['tag'])

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(sentences)
```

```python
model = LogisticRegression()
model.fit(X, labels)
```

```python
def chatbot_response(user_input):
    user_input_vect = vectorizer.transform([user_input])
    intent = model.predict(user_input_vect)[0]

    for i in data['intents']:
        if i['tag'] == intent:
            return random.choice(i['responses'])

# Test chatbot
while True:
    inp = input("You: ")
    if inp.lower() == "quit":
        break
    print("Bot:", chatbot_response(inp))
```
